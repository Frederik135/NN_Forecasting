```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense, RepeatVector, TimeDistributed
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from keras.models import Model
import seaborn as sns
import yfinance as yf

def fetch_stock_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data

ticker = 'GOOGL'
end_date = pd.Timestamp.today()
start_date = end_date - pd.DateOffset(years=5)
data = fetch_stock_data(ticker, start_date, end_date)

# Reset the index to get 'Date' as a column
df = data.reset_index()
# Now you can directly access 'Close' and 'Date'
df['Date'] = pd.to_datetime(df['Date'])

sns.lineplot(x=df['Date'], y=df['Close'])
plt.show()
print("Start date is: ", df['Date'].min())
print("End date is: ", df['Date'].max())


train, test = df.loc[df['Date'] <= '2022-01-01'], df.loc[df['Date'] > '2022-01-01']

scaler = StandardScaler()
scaler = scaler.fit(train[['Close']])

train['Close'] = scaler.transform(train[['Close']])
test['Close'] = scaler.transform(test[['Close']])


# Set a practical sequence size
seq_size = 10  # For example, use the last 10 days to predict the next day

# Adjusted to_sequences function to handle a reasonable sequence size
def to_sequences(x, y, seq_size=1):
    x_values = []
    y_values = []

    for i in range(len(x) - seq_size):
        x_values.append(x.iloc[i:(i + seq_size)].values)
        y_values.append(y.iloc[i + seq_size])
        
    return np.array(x_values), np.array(y_values)

# Generate sequences with the new seq_size
trainX, trainY = to_sequences(train[['Close']], train['Close'], seq_size)
testX, testY = to_sequences(test[['Close']], test['Close'], seq_size)

# Ensure that you have the data in the correct shape for LSTM
print("Shape of trainX:", trainX.shape)
print("Shape of trainY:", trainY.shape)

# Proceed with model creation and training
model = Sequential()
model.add(LSTM(128, input_shape=(seq_size, 1)))  # Adjust input_shape to (seq_size, 1)
model.add(Dropout(rate=0.2))
model.add(RepeatVector(seq_size))
model.add(LSTM(128, return_sequences=True))
model.add(Dropout(rate=0.2))
model.add(TimeDistributed(Dense(1)))  # Assuming you're predicting a single value
model.compile(optimizer='adam', loss='mae')
model.summary()

history = model.fit(trainX, trainY, epochs=10, batch_size=32, validation_split=0.1, verbose=1)

plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()

trainPredict = model.predict(trainX)
trainMAE = np.mean(np.abs(trainPredict - trainX), axis=1)
plt.hist(trainMAE, bins=30)
max_trainMAE = 0.3

testPredict = model.predict(testX)
testMAE = np.mean(np.abs(testPredict - testX), axis=1)
plt.hist(testMAE, bins=30)

anomaly_df = pd.DataFrame(test[seq_size:])
anomaly_df['testMAE'] = testMAE
anomaly_df['max_trainMAE'] = max_trainMAE
anomaly_df['anomaly'] = anomaly_df['testMAE'] > anomaly_df['max_trainMAE']
anomaly_df['Close'] = test[seq_size:]['Close']

sns.lineplot(x=anomaly_df['Date'], y=anomaly_df['testMAE'])
sns.lineplot(x=anomaly_df['Date'], y=anomaly_df['max_trainMAE'])

# Ensure that the data is in the correct shape for inverse transformation
# Convert 'Close' column to a 2D array with .values.reshape(-1, 1)
close_prices_reshaped = anomaly_df['Close'].values.reshape(-1, 1)
anomaly_close_prices_reshaped = anomalies['Close'].values.reshape(-1, 1)

# Apply the inverse transformation
anomaly_df['Close_inversed'] = scaler.inverse_transform(close_prices_reshaped)
anomalies['Close_inversed'] = scaler.inverse_transform(anomaly_close_prices_reshaped)

# Now plot using the inversely transformed 'Close' prices
sns.lineplot(x=anomaly_df['Date'], y=anomaly_df['Close_inversed'])
sns.scatterplot(x=anomalies['Date'], y=anomalies['Close_inversed'], color='r')
plt.show()
          
```
